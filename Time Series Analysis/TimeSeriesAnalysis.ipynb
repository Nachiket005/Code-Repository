### Time Series Decomposition

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

### Read AirPassenger Data

```python
import pandas as pd 
df = pd.read_csv('AirPassengers.csv')
df.head()
```

```python
df.info()
```

```python
df.columns = ['Month', 'nPass']
df.head()
```

### Convert month column to datetime

```python
df['Month'] = pd.to_datetime(df['Month'])
df.head()
```

```python
df.dtypes
```

```python
df.index = df['Month']
```

```python
df
```

### Drop month column

```python
df = df.drop(columns=['Month'])
df.head()
```

### Visualize the time series data

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,10))
plt.plot(df['nPass'])
plt.title('Time data series for AirPassengers')
plt.xlabel('Date')
plt.ylabel('No. of Passengers')
plt.show()
```

### Time data Decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
T = seasonal_decompose(df['nPass'], model='multiplicative')
```

```python
T.plot()
```

```python
T.trend.plot()
```

```python
T.seasonal.plot()
```

```python
T.resid.plot()
```

### Restart and run all cells

```python
from pmdarima import auto_arima
```

```python
model = auto_arima(df['nPass'], seasonal=True, m=12)
```

```python
model.order
```

### Model selected is ARIMA(p=2, d=1, q=1)

```python
forecast, conf_interval = model.predict(n_periods=36, return_conf_int=True)
```

```python
forecast
```

```python
forecast = forecast.apply(round)
```

```python
forecast
```

```python
conf_interval
```

```python
lower = conf_interval[:,0]
upper = conf_interval[:,1]
```

```python
upper
```

```python
lower
```

### Plot the forecast along with original data

```python
plt.figure(figsize=(16,10))
plt.plot(df['nPass'], label='original')
plt.plot(forecast, label='forecast')
plt.fill_between(forecast.index, lower, upper, alpha=0.5, color='gray', label='confidance interval')
```

### Save forecast data into dataframe

```python
df_pred = forecast.to_frame()
df_pred
```

```python
df_pred.columns = ['Forecast']
```

```python
df_pred
```

```python
df_pred['lower'] = lower
df_pred['upper'] = upper
```

```python
df_pred
```

```python
df_pred['lower'] = df_pred['lower'].apply(round)
df_pred['upper'] = df_pred['upper'].apply(round)
```

```python
df_pred
```

### Save predictions to csv

```python
df_pred.to_csv('Predictions.csv')
```
